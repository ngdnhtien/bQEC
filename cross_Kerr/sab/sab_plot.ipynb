```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_formats = ['svg']
```

```python
AqcB = np.load('SAB_plots_AqcB.npy')
AqcBqb_qb0 = np.load('SAB_plots_AqcBqb_qb=0.npy')
AqcBqb_qc009 = np.load('SAB_plots_AqcBqb_qc=0.09.npy')
```

```python
from matplotlib.ticker import ScalarFormatter
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(4*2.7, 4))

MHz = 1e6
GHz = 1e9

ax1.plot(AqcBqb_qb0[0], AqcBqb_qb0[1]*GHz/((2*np.pi)*MHz), marker='o', alpha=0.7, label=r'self-Kerr $K_A$', color='red')
ax1.plot(AqcBqb_qb0[0], AqcBqb_qb0[2]*GHz/((2*np.pi)*MHz), marker='o', alpha=0.7, label=r'self-Kerr $K_B$', color='blue')
ax1.plot(AqcBqb_qb0[0], AqcBqb_qb0[3]*GHz/((2*np.pi)*MHz), marker='o', alpha=0.7, label=r'cross-Kerr $K_{AB}$', color='green')
ax1.axhline(0, linestyle='-', color='grey')
ax1.tick_params(axis='both', which='major', labelsize=15)
ax1.set_ylim([min(AqcBqb_qb0[1]*GHz/((2*np.pi)*MHz))-0.15e-2, 1.45e-2])
ax1.set_xlabel(r'Drive amplitude $|\Omega_{dqc}/\delta_{dqc}|^2$', fontsize=15)
ax1.set_ylabel(r'Cavity nonlinearities (MHz)', fontsize=15)

ax2.plot(AqcBqb_qc009[0], AqcBqb_qc009[1]*GHz/((2*np.pi)*MHz), marker='o', alpha=0.7, color='red')
ax2.scatter(AqcBqb_qc009[0], AqcBqb_qc009[2]*GHz/((2*np.pi)*MHz), marker='o', alpha=0.7, color='blue')
ax2.plot(AqcBqb_qc009[0][0:4], AqcBqb_qc009[2][0:4]*GHz/((2*np.pi)*MHz), marker='o', alpha=0.7, color='blue')
ax2.plot(AqcBqb_qc009[0][5:], AqcBqb_qc009[2][5:]*GHz/((2*np.pi)*MHz), marker='o', alpha=0.7, color='blue')
ax2.plot(AqcBqb_qc009[0], AqcBqb_qc009[3]*GHz/((2*np.pi)*MHz), marker='o', alpha=0.7, color='green')
ax2.axhline(0, linestyle='-', color='grey')
ax2.axvline(0.7*1e-1, linestyle='--', color='blue')
ax2.tick_params(axis='both', which='major', labelsize=15)
ax2.set_ylim([min(AqcBqb_qc009[2]*GHz/((2*np.pi)*MHz))-0.15e-2, 0.6*1e-1])
ax2.set_xlabel(r'Drive amplitude $|\Omega_{dqb}/\delta_{dqb}|^2$', fontsize=15)
ax2.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax2.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

for ax in [ax1, ax2]:
    ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

handles, labels = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
handles += handles2
labels += labels2

fig.legend(handles, labels, loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1.05), fontsize=13)

fig.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
fig.savefig('plot_transparent.png', dpi=600, transparent=True, bbox_inches='tight')
```
