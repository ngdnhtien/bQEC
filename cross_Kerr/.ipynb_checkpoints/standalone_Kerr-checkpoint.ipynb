```python
from scipy.optimize import curve_fit

def fit_function(x_values, y_values, function, init_params):
    """fit_function _summary_

    _extended_summary_

    Args:
        x_values (_type_): _description_
        y_values (_type_): _description_
        function (_type_): _description_
        init_params (_type_): _description_

    Returns:
        _type_: _description_
    """
    fitparams, _ = curve_fit(function, x_values, y_values, init_params)
    y_fit = function(x_values, *fitparams)

    return fitparams, y_fit
```

## Bare system

```python
# Transmon mode C
omega_10 = 4.936 # GHz
alpha = 0.168 # GHz
omega_c = omega_10 + alpha

# Ratios
r1 = 9.64
r2 = 0.064

# Cavity A 
delta_a = r1 * alpha
g_a = r2 * delta_a
omega_a = omega_10 + delta_a

# Drive parameters
r3 = 0.08
delta_d = r3 * alpha
omega_d = omega_10 + delta_d

r4 = 1
Omega_d = r4 * delta_d

r5 = 1
delta_da = (r5*omega_d - omega_a)
delta_dc = (r5*omega_d - omega_c)
```

```python
H_cav = - delta_da * a.dag() * a
H_trm = - delta_dc * c.dag() * c - (alpha / 2) * (c.dag() * c + 1) * c.dag() * c
H_drv = Omega_d * c.dag() + conj(Omega_d) * c
H_int = g_a * a * c.dag() + conj(g_a) * a.dag() * c
```

## Bare Hamiltonian

```python
H_tm_bare_RWA = H_cav + H_trm 
```

```python
bare_tm_eig_energies, bare_tm_eig_states = H_tm_bare_RWA.eigenstates()
```

```python
bare_labelling = []

for idx, state in enumerate(bare_tm_eig_states):
    n_a_val = np.real((state.dag() * n_a * state).full()[0][0])
    n_c_val = np.real((state.dag() * n_c * state).full()[0][0])

    bare_labelling.append((n_a_val, n_c_val))

bare_labelling = np.array(bare_labelling)
```

```python
switchCoupling = 1
switchDrive = 1

H_tm_entangled_RWA = H_cav + H_trm + switchCoupling * H_int + switchDrive * H_drv
```

## Reproducing two-mode results

```python
def comp_self_Kerr(r4, dim_a, dim_c):
    
    a = tensor(destroy(dim_a), qeye(dim_c))
    c = tensor(qeye(dim_a), destroy(dim_c))

    n_a = a.dag() * a
    n_c = c.dag() * c
    
    # Transmon mode C
    omega_10 = 4.936 # GHz
    alpha = 0.168 # GHz
    omega_c = omega_10 + alpha

    # Ratios
    r1 = 9.64
    r2 = 0.064
    r3 = 0.08

    # Cavity A 
    delta_a = r1 * alpha
    g_a = r2 * delta_a
    omega_a = omega_10 + delta_a

    # Drive parameters
    delta_d = r3 * alpha
    omega_d = omega_10 + delta_d

    Omega_d = r4 * delta_d

    delta_da = (omega_d - omega_a)
    delta_dc = (omega_d - omega_c)
    
    H_cav = - delta_da * a.dag() * a
    H_trm = - delta_dc * c.dag() * c - (alpha / 2) * (c.dag() * c + 1) * c.dag() * c
    H_drv = Omega_d * c.dag() + conj(Omega_d) * c
    H_int = g_a * a * c.dag() + conj(g_a) * a.dag() * c
    
    H_tm_bare_RWA = H_cav + H_trm 
    
    bare_tm_eig_energies, bare_tm_eig_states = H_tm_bare_RWA.eigenstates()
    
    bare_labelling = []

    for idx, state in enumerate(bare_tm_eig_states):
        n_a_val = np.real((state.dag() * n_a * state).full()[0][0])
        n_c_val = np.real((state.dag() * n_c * state).full()[0][0])

        bare_labelling.append((n_a_val, n_c_val))

    bare_labelling = np.array(bare_labelling)
    
    switchCoupling = 1
    switchDrive = 1

    H_tm_entangled_RWA = H_cav + H_trm + switchCoupling * H_int + switchDrive * H_drv
    
    entangled_tm_eig_energies, entangled_tm_eig_states = H_tm_entangled_RWA.eigenstates()
    
    entangled_labelling = bare_labelling-bare_labelling

    for i, entangled_state in enumerate(entangled_tm_eig_states):

        highest_fidelity_found = 0
        idx_found_by_state_fidelity = 0

        for j, uncoupled_state in enumerate(bare_tm_eig_states):

            # Compute state fidelity
            state_fidelity = fidelity(entangled_state, uncoupled_state)

            # Update index
            if state_fidelity > highest_fidelity_found:
                highest_fidelity_found = state_fidelity
                idx_found_by_state_fidelity = j

        # Acquire entangled labelling index
        entangled_labelling[i] = bare_labelling[idx_found_by_state_fidelity]
        
    cond = entangled_labelling[:, 1] == 0
    transmon_ground_state_indices = np.where(cond)[0]
    entangled_labelling_filtered = entangled_labelling[transmon_ground_state_indices]
    
    NA = entangled_labelling_filtered[:, 0]
    Energies = entangled_tm_eig_energies[transmon_ground_state_indices]

    Energies = Energies - Energies[0]
    
    parameters, fit = fit_function(
        NA[:-1], Energies[:-1], 
        lambda x, c1, c2, c3, c4, c5: ((-delta_da+c1) * x + c2*x**2 + c3*x**3 + c4*x**4 + c5*x**5),
        [0, 0, 0, 0, 0]
    )

    return parameters[1]*2, parameters[2]*np.math.factorial(3), parameters[3]*np.math.factorial(4)
```

```python
def energy_level(r2, dim_a, dim_c):
    
    a = tensor(destroy(dim_a), qeye(dim_c))
    c = tensor(qeye(dim_a), destroy(dim_c))

    n_a = a.dag() * a
    n_c = c.dag() * c
    
    # Transmon mode C
    omega_10 = 4.936 # GHz
    alpha = 0.168 # GHz
    omega_c = omega_10 + alpha

    # Ratios
    r1 = 9.64

    # Cavity A 
    delta_a = r1 * alpha
    g_a = r2 * delta_a
    omega_a = omega_10 + delta_a

    # Drive parameters
    r3 = 0.08
    delta_d = r3 * alpha
    omega_d = omega_10 + delta_d

    r4 = 0.5
    Omega_d = r4 * delta_d
     
    r5 = 1 # Driving frame
    delta_da = (r5 * omega_d - omega_a)
    delta_dc = (r5 * omega_d - omega_c)
    
    H_cav = - delta_da * a.dag() * a
    H_trm = - delta_dc * c.dag() * c - (alpha / 2) * (c.dag() * c + 1) * c.dag() * c
    H_drv = Omega_d * c.dag() + conj(Omega_d) * c
    H_int = g_a * a * c.dag() + conj(g_a) * a.dag() * c

    H = H_cav + H_trm + H_int
    
    eig_energy, _ = H.eigenstates()
    
    return eig_energy[-3], eig_energy[-2], eig_energy[-1]    
```

## Driven system

```python
def Hamiltonian(r4, couplingOn, driveOn):
    
    # Transmon mode C
    omega_10 = 4.936 # GHz
    alpha = 0.168 # GHz
    omega_c = omega_10 + alpha

    # Ratios
    r1 = 9.64 
    r2 = 0.064

    # Cavity A 
    delta_a = r1 * alpha
    g_a = r2 * delta_a
    omega_a = omega_10 + delta_a

    # Drive parameters
    r3 = 0.08
    delta_d = 0.08 * alpha
    omega_d = omega_10 + delta_d

    Omega_d = r4 * delta_d

    delta_da = (omega_d - omega_a)
    delta_dc = (omega_d - omega_c)

    H_cav = - delta_da * a.dag() * a
    H_trm = - delta_dc * c.dag() * c - (alpha / 2) * (c.dag() * c + 1) * c.dag() * c
    H_drv = Omega_d * c.dag() + conj(Omega_d) * c
    H_int = g_a * a * c.dag() + conj(g_a) * a.dag() * c
    
    switchCoupling = 0
    switchDrive = 0
    
    if couplingOn == True:
        switchCoupling = 1
    if driveOn == True:
        switchDrive = 1
        
    return H_cav + H_trm + switchCoupling * H_int + switchDrive * H_drv
```
