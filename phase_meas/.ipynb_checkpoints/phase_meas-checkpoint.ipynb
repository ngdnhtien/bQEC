# Non-normalizable phase states

I mean when $N\to \infty$...

```python
def phase_state(dim_Fock, phi):
    '''
    Generate a phase state
    '''
    state = 0
    for n in range(dim_Fock):
        state += np.exp(1j * n *phi) * qt.fock(dim_Fock, n)
    state = (1 / (np.sqrt(2 * np.pi))) * state
    
    return state
```

```python
dim_Fock = 500
phi = 0

phase_zero = phase_state(dim_Fock, phi).unit()
```

```python
def a_very_slowly_decreasing_state_wrt_n(b, dimFock):
    '''
    Generating a very slowly decrasing state
    '''
    
    if abs(b) > 1:
        the_state = 0
        for n in range(dimFock):
            the_state += (1 / (b**n)) * fock(dimFock, n)
        the_state = sqrt((b**2 - 1)/b**2) * the_state
        
        return the_state
    
    else: 
        raise Exception("Non non! Je veut une abs(b) > 1, si vous plait!")
```

## Verify the reduction

```python
def nat_fact(integer):
    nat_fact = 0
    for i in range(1, integer+1):
        nat_fact += ln(i)
        
    return nat_fact
```

```python
def lnS_l(l, alpha, N):
    '''
        Natural logarithm of the sum over l
    '''
    
    A = N * (l + 1)
    B = l * N
    
    lnS_l = (l + 1/2) * N * ln(alpha) - (1 / 2) * nat_fact(A) - (1 / 2) * nat_fact(B)
    
    return lnS_l
```

```python
infinities = [10, 100, 500, 1000, 2000, 5000]
```

```python
N = 3
infinity = infinities[1]
l_range = np.arange(0, infinity, 1)

lnS_list = []
for alpha in np.linspace(1, 50, 100):
    
    lnS = 0

    for l in l_range:
        lnS += exp(lnS_l(l, alpha, N))

    lnS_list.append(lnS)
```

```python
plt.plot(np.linspace(1, 100, 100), ln(lnS_list))
```
